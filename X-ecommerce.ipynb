```python
x = pd.merge(x_order_report , x_sku_master , on = 'SKU')
x.head()
```

```python
x.rename(columns = {'Weight (g)' : 'weight_x'} , inplace = True)
x.rename(columns= {'ExternOrderNo' : 'Order ID'} , inplace = True)
x['weight_x']  = x['weight_x']/1000
def weight_slab_x(value):
    if value<=0.5:
        return 0.5
    elif 0.5 <value <=1:
        return 1
    elif 1<value<=1.5:
        return 1.5
    elif 1.5 <value <= 2:
        return 2
    elif 2< value <= 2.5:
        return 2.5
    elif 2.5< value<= 3:
        return 3
    elif 3< value <= 3.5:
        return 3.5
    elif 3.5< value<= 4:
        return 4
x['weight_slab_x'] = x['weight_x'].map(weight_slab_x)
```

```python
x_pincode.rename(columns = {'Zone' : 'Zone_x'} , inplace = True)
x_pincode.head()
```

```python
company = pd.merge(courier_company , x_pincode , on = 'Customer Pincode')
company.head()
```

```python
final = pd.merge(company , x, on = 'Order ID')
final.rename(columns= {'Zone' : 'Zone_c'} , inplace = True)
final.head()
```

```python
def weight_slab_c(value):
    if value<=0.5:
        return 0.5
    elif 0.5 <value <=1:
        return 1
    elif 1<value<=1.5:
        return 1.5
    elif 1.5 <value <= 2:
        return 2
    elif 2< value <= 2.5:
        return 2.5
    elif 2.5< value<= 3:
        return 3
    elif 3< value <= 3.5:
        return 3.5
    elif 3.5< value<= 4:
        return 4
final['weight_slab_c'] = final['Charged Weight'].map(weight_slab_c)
final['weight_slab_x'] = final['weight_slab_x'] * final['Order Qty']
final.head()
```

```python
c_rates
```

```python
x = final['weight_slab_x']
y = final['Zone_x']
z = final['Type of Shipment']
def func(a,b,c):
    new = []
    for i in zip(a,b,c):
        ########## for zone a ################
        if i[1] == 'a' and i[2] == 'Forward charges' and i[0] <= 0.5:
            new.append(29.5) 
        elif i[1] == 'a' and i[-1] == 'Forward charges' and i[0] > 0.5:
            new.append(29.5 + (i[0]/0.5 -1) * 23.6)
            
        elif i[1] == 'a' and i[2] == 'Forward and RTO charges' and i[0] <= 0.5:
            new.append(29.5 + 13.6)
        elif i[1] == 'a' and i[-1] == 'Forward and RTO charges' and i[0] > 0.5:
            new.append(29.5 + 13.6 + ((i[0]/0.5 -1) * 23.6)  + (i[0]/0.5 -1) * 23.6 )
        
        ################# For zone b #######################
        elif i[1] == 'b' and i[2] == 'Forward charges' and i[0] <= 0.5:
            new.append(33) 
        elif i[1] == 'b' and i[-1] == 'Forward charges' and i[0] > 0.5:
            new.append(33 + (i[0]/0.5 -1) * 28.3 )
            
        elif i[1] == 'b' and i[2] == 'Forward and RTO charges' and i[0] <= 0.5:
            new.append(33 + 20.5)
        elif i[1] == 'b' and i[-1] == 'Forward and RTO charges' and i[0] > 0.5:
            new.append(33 + 20.5 + ((i[0]/0.5 -1) * 28.3)  + (i[0]/0.5 -1) * 28.3 )  
            
    ####################### For zone c #############################   
    
        elif i[1] == 'c' and i[2] == 'Forward charges' and i[0] <= 0.5:
            new.append(40.1) 
        elif i[1] == 'c' and i[-1] == 'Forward charges' and i[0] > 0.5:
            new.append(40.1 + (i[0]/0.5 -1) * 38.9)
            
        elif i[1] == 'c' and i[2] == 'Forward and RTO charges' and i[0] <= 0.5:
            new.append(40.1 + 31.9)
        elif i[1] == 'c' and i[-1] == 'Forward and RTO charges' and i[0] > 0.5:
            new.append(40.1 + 31.9 + ((i[0]/0.5 -1) * 38.9)  + (i[0]/0.5 -1) * 38.9) 
    
    ##################### For zone d ################################
    
        elif i[1] == 'd' and i[2] == 'Forward charges' and i[0] <= 0.5:
            new.append(45.4) 
        elif i[1] == 'd' and i[-1] == 'Forward charges' and i[0] > 0.5:
            new.append(45.4 + (i[0]/0.5 -1) * 44.8)
            
        elif i[1] == 'd' and i[2] == 'Forward and RTO charges' and i[0] <= 0.5:
            new.append(45.4 + 41.3)
        elif i[1] == 'd' and i[-1] == 'Forward and RTO charges' and i[0] > 0.5:
            new.append(45.4 + 41.3 + ((i[0]/0.5 -1) * 44.8)  + (i[0]/0.5 -1) * 44.8)
            
    ##################### For zone e ################################
    
        elif i[1] == 'e' and i[2] == 'Forward charges' and i[0] <= 0.5:
            new.append(56.6) 
        elif i[1] == 'e' and i[-1] == 'Forward charges' and i[0] > 0.5:
            new.append(56.6 + (i[0]/0.5 -1) * 55.5)
            
        elif i[1] == 'e' and i[2] == 'Forward and RTO charges' and i[0] <= 0.5:
            new.append(56.6 + 50.7)
        elif i[1] == 'e' and i[-1] == 'Forward and RTO charges' and i[0] > 0.5:
            new.append(56.6 + 50.7 + ((i[0]/0.5 -1) * 55.5)  + (i[0]/0.5 -1) * 55.5)
    return new
charges  = func(x,y,z)

final['Expected Charge as per X (Rs.)'] = charges * final['Order Qty']
final.drop('Type of Shipment' , axis = 1 , inplace = True)
final.head()
```

```python
final.head()
```

```python
final ['difference_charge'] = final['Billing Amount (Rs.)'] - final['Expected Charge as per X (Rs.)']
final.head()
```

```python
final = final[['AWB Code' ,'weight_x' , 'weight_slab_x' , 'Charged Weight' , 'weight_slab_c' , 
       'Zone_x' , 'Zone_c' , 'Expected Charge as per X (Rs.)' , 'Billing Amount (Rs.)' , 
               'difference_charge']]
final.head()
```

```python
diff = final['difference_charge']
c_charge = final['Expected Charge as per X (Rs.)']
def output(x,y):  
    count_c = 0
    count_u = 0
    count_o = 0
    sum_c = 0
    sum_u = 0
    sum_o = 0
    for i,j in zip(x, y):
        if i ==0 :
            count_c+=1
            sum_c+=j
        elif i>0:
            count_o+=1
            sum_o+=i
        elif i<0:
            count_u+=1
            sum_u+=abs(i)
    return (count_c , count_u, count_o) , (sum_c,sum_u,sum_o)

a,b = output(diff , c_charge)
w = ['Total orders where X has been correctly charged' , 
     'Total Orders where X has been overcharged' , 'Total Orders where X has been undercharged']
```

```python
df = pd.DataFrame({
    'output' : w,
    'count' : a,
    'Amount (Rs.)' : b
})
df
```
